## Pipeline for Data Augumentation

```python
def aug_flow(img1):
    data = []
    for i in range(9):
        img = tf.image.random_brightness(img1, max_delta=0.02)
        img = tf.image.random_contrast(img1, lower=0.6, upper=1)
        img = tf.image.random_flip_left_right(img1)
        img = tf.image.random_jpeg_quality(img1, min_jpeg_quality=90, max_jpeg_quality=100)
        img = tf.image.random_saturation(img1, lower=0.9,upper=1)

        data.append(img1)
    return data
```

```python
augumented_images=aug_flow(img1)
```

```python

for idx, aug_img in enumerate(augumented_images):
    plt.subplot(3,3,idx+1)
    bgr = cv2.cvtColor(aug_img, cv2.COLOR_BGR2RGB)
    plt.imshow(bgr)
    
```
